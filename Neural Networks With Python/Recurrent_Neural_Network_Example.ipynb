```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
import matplotlib.pyplot as plt

# Sample text data
data = """
Once upon a time, in the heart of the quaint village of Willowbrook, there stood a grocery store unlike any other. This wasn't just any grocery store; it was known to the villagers as "The Enchanted Grocer." Its weathered sign hung above the door, swaying gently in the breeze, with the words painted in elegant, curling letters.

The store had been there for as long as anyone could remember. Some said it was as old as the village itself, while others whispered that it had appeared one foggy morning long ago, brought to life by some ancient magic. The owner of this mysterious establishment was an elderly man named Mr. Jasper Whittle. With his silver hair and twinkling blue eyes, Mr. Whittle seemed to embody the very essence of the store – old, kind, and full of secrets.

Mr. Whittle's grocery store was a treasure trove of wonders. Shelves upon shelves lined the walls, filled with jars of herbs, spices, and preserves that shimmered with an otherworldly glow. The fruits and vegetables, freshly harvested from the surrounding farms, were always ripe and bursting with color. It was said that the produce in The Enchanted Grocer never spoiled, and indeed, the apples always had a crisp snap, the tomatoes were juicy and sweet, and the carrots were as orange as a sunset.

Every corner of the store was filled with surprises. There was the aisle of teas that promised to cure any ailment or lift any mood. Each tin was labeled not just with the type of tea, but with its magical properties: "Peppermint for Pleasant Dreams," "Chamomile for Calming Storms," and "Elderflower for Extra Energy." In another section, there were shelves of freshly baked bread, so warm and fragrant that the scent alone could make one's mouth water. The bread, it was rumored, had the power to evoke fond memories of home with every bite.

One of the store's most enchanting features was the enchanted door at the back, which led to the "Room of Wishes." This room was a small, cozy nook filled with twinkling fairy lights, where a large, ancient oak tree grew right through the middle of the floor. Its branches spread out to form a canopy over the room, and hanging from the branches were tiny glass bottles. Each bottle contained a piece of parchment on which customers could write their wishes. Mr. Whittle always reminded visitors that while the wishes often came true, they worked in mysterious ways and sometimes required a bit of patience and a lot of faith.

The villagers loved to visit The Enchanted Grocer not just for the unique items it sold, but also for the sense of wonder and community it fostered. Mr. Whittle was always ready with a kind word, a bit of sage advice, or a fascinating story. Children would gather around him, wide-eyed, as he spun tales of the store's history and the magical creatures that helped tend the garden out back.

One particularly memorable story Mr. Whittle loved to tell was about the time the store's magic saved the village from a terrible drought. It had been a scorching summer, and the fields were dry and cracked, the riverbeds empty. The villagers were on the brink of despair when one evening, a mysterious fog rolled into Willowbrook, and the air was filled with the scent of rain. Mr. Whittle, with his knowing smile, had led the villagers to the store, where they found the enchanted door wide open, revealing the Room of Wishes bathed in a golden light. The villagers hurriedly wrote their wishes for rain and hung the bottles on the oak tree. That very night, a gentle rain began to fall, revitalizing the fields and filling the rivers, saving the village from disaster.

One day, a young woman named Eliza arrived in Willowbrook. She was a traveler with no particular destination, drawn to the village by stories of its enchanting grocery store. Eliza was curious and had an open heart, always ready for an adventure. When she entered The Enchanted Grocer for the first time, she felt an immediate sense of belonging, as if the store itself welcomed her with open arms.

Eliza quickly became a regular visitor, helping Mr. Whittle with the daily tasks of running the store. She learned about the magical properties of the items they sold and listened intently to Mr. Whittle's stories. One day, as she was organizing the shelves, she found an old, dusty book hidden behind a row of flour sacks. The book was bound in worn leather and filled with handwritten notes and sketches. It was Mr. Whittle's journal, chronicling the history of the store and his life as its caretaker.

Reading the journal, Eliza discovered that Mr. Whittle was not just a grocer but a guardian of ancient magic. The store was a living entity, infused with enchantments that protected and nurtured the village. Mr. Whittle had been chosen as its caretaker many years ago, and now, as he grew older, he was searching for someone to take his place.

One evening, as the sun set and the village was bathed in golden light, Mr. Whittle took Eliza aside. He told her about the journal and his hope that she would become the next guardian of The Enchanted Grocer. Eliza was taken aback but honored. She had come to love the store and the village, and she felt a deep connection to the magic that flowed through the place.

With Mr. Whittle's guidance, Eliza began her training. She learned to tend the enchanted garden, to mix the special teas, and to listen to the whispers of the ancient oak tree in the Room of Wishes. As she grew into her new role, the villagers welcomed her with open arms, grateful that the magic of The Enchanted Grocer would continue to thrive.

Years passed, and Mr. Whittle, now frail but content, watched with pride as Eliza took over the store. Under her care, The Enchanted Grocer flourished, becoming even more vibrant and full of wonder. The villagers continued to visit, finding comfort and joy in the magical store that had always been a beacon of hope and community.

And so, the legacy of The Enchanted Grocer lived on, a timeless tale of magic, kindness, and the enduring power of community. Eliza, with her warm heart and adventurous spirit, ensured that the store remained a place where dreams came true, and the magic of Willowbrook would never fade.
"""

# Prepare the tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
total_words = len(tokenizer.word_index) + 1

# Create input sequences using list of tokens
input_sequences = []
for line in data.split('.'):
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

# Create predictors and label
X, y = input_sequences[:,:-1], input_sequences[:,-1]
y = tf.keras.utils.to_categorical(y, num_classes=total_words)

# Define the model
model = Sequential()
model.add(Embedding(total_words, 100, input_length=max_sequence_len-1))
model.add(LSTM(150))
model.add(Dense(total_words, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model and capture the training history
history = model.fit(X, y, epochs=100, verbose=1)

# Plot training loss and accuracy
plt.figure(figsize=(14, 5))

# Plot training loss
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper right')

# Plot training accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train'], loc='upper left')

plt.show()

# Function to generate text
def generate_text(seed_text, next_words, max_sequence_len):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)
        output_word = tokenizer.index_word[predicted[0]]
        seed_text += " " + output_word
    return seed_text

# Generate random text
seed_text = "Once upon a time"
next_words = 100
print(generate_text(seed_text, next_words, max_sequence_len))
```
