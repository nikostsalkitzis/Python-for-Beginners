```python
from sklearn.datasets import load_wine
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# Load the Wine dataset from scikit-learn
data = load_wine()

# Convert the data to a DataFrame
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add a target column to the DataFrame
df['target'] = data.target

# Optionally, you can also add the target_names as a column
df['target_names'] = [data.target_names[i] for i in data.target]

# Print the first few rows of the DataFrame
print(df.head())

#Lets see some statistical info about our data
print(df.info())
print(df.describe())
print(df['target'].value_counts())
print(df.isnull().sum())
#we can see that we have three classes of wines
#data visualization
def histogram_plotting(df):
    num_cols = 3  # Number of columns in the subplot grid
    num_rows = (len(df.columns) + num_cols - 1) // num_cols

    # Create subplots with histograms for each column
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))  # Adjust the figure size as needed

    for i, column in enumerate(df.columns):
        row = i // num_cols
        col = i % num_cols
        ax = axes[row, col]

        df[column].hist(ax=ax, bins=20)  # You can adjust the number of bins as needed
        ax.set_title(f'Histogram of {column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Frequency')

    # Remove any empty subplots
    if i < (num_rows * num_cols) - 1:
        for j in range(i + 1, num_rows * num_cols):
            fig.delaxes(axes.ravel()[j])

    plt.tight_layout()
    plt.show()
def density_plotting(df):
    # Set the style of the plots (optional)
    sns.set(style="whitegrid")

    # Get the list of column names
    column_names = df.columns

    # Calculate the number of rows and columns for the subplot grid
    num_cols = 3  # Number of columns in the subplot grid
    num_rows = (len(column_names) + num_cols - 1) // num_cols

    # Create density plots for each column using subplots
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))  # Adjust the figure size as needed

    for i, column in enumerate(column_names):
        row = i // num_cols
        col = i % num_cols
        ax = axes[row, col]

        sns.kdeplot(df[column], shade=True, ax=ax)
        ax.set_title(f'Density Plot of {column}')
        ax.set_xlabel(column)
        ax.set_ylabel('Density')

    # Remove any empty subplots
    if i < (num_rows * num_cols) - 1:
        for j in range(i + 1, num_rows * num_cols):
            fig.delaxes(axes.ravel()[j])

    plt.tight_layout()
    plt.show()


def create_boxplots(dataframe):

    # Create a new figure
    plt.figure(figsize=(12, 6))

    # Loop through each column and create a box plot
    for column in dataframe.columns:
        plt.boxplot(dataframe[column], vert=False)
        plt.title(f'Box Plot of {column}')
        plt.xlabel(column)
        plt.ylabel('Values')
        plt.show()

def create_pair_plots(df):
    sns.set(style="ticks")
    sns.pairplot(df, diag_kind="kde")
    plt.show()
def feature_correlation(df):
    correlation = df.corr()['target'].abs()
    correlation_sorted = correlation.sort_values(ascending=False)

    plt.figure(figsize=(10, 6), dpi=300)

    bar_colors = ['#446A46', '#E4AEC5']

    correlation_sorted.plot(kind='bar', color=bar_colors)
    plt.title('Feature Importance (Correlation)', color='#446A46')
    plt.xlabel('Features', color='#446A46')
    plt.ylabel('Correlation', color='#446A46')
    plt.xticks(color='#446A46')
    plt.yticks(color='#446A46')
    plt.show()
```


Histograms are a type of data visualization that provides insight into the distribution of a continuous variable. They display the frequencies of data points falling into different "bins" or intervals. Histograms are particularly useful for understanding the underlying patterns, trends, and characteristics of a dataset.

Here's what histograms show and how they work:

Distribution of Data: A histogram shows how data points are distributed across various ranges or intervals. Each interval, or bin, on the x-axis represents a specific range of values.

Frequency or Count: The height of each bar in the histogram corresponds to the frequency or count of data points within the corresponding bin. In other words, the y-axis represents the number of occurrences of data points within that range.

Shape of the Distribution: The overall shape of the histogram provides insights into the underlying distribution of the data. Common distribution shapes include normal (bell-shaped), skewed (asymmetric), bimodal (having two peaks), and more.

Central Tendency: The center of the distribution can often be determined by locating the bin with the highest frequency, which corresponds to the mode (most frequent value). In symmetric distributions, the mode is also the mean and median.

Spread and Dispersion: The width of the histogram bins and the distribution's spread can give you an idea of the variability or dispersion of the data.

Outliers: Outliers (extreme values) can sometimes be identified as data points that fall outside the typical range of the histogram.

Skewness: Skewness, which indicates the asymmetry of the distribution, can be observed by looking at how the data is distributed on either side of the center.

Gaps and Clusters: Gaps and clusters of data in certain intervals can reveal patterns or insights about the data.

Histograms are widely used in data analysis and statistics to explore and visualize the basic characteristics of a continuous variable. They are also often used as a precursor to more advanced statistical analysis, like fitting distributions or making hypotheses about the data's behavior.

A density plot, also known as a kernel density plot, is a type of data visualization that provides a smoothed representation of the distribution of a continuous variable. Unlike histograms, which use discrete bins to approximate the distribution, density plots use a continuous curve to estimate the probability density function of the data. Density plots offer a more continuous and smooth representation of the data's distribution.

Here's what density plots show and how they work:

Smoothed Distribution: Density plots display the smoothed probability density function of the data. This means that the curve represents how the data is likely to be distributed across the entire range of values, rather than just within specific bins.

Continuous Curve: Unlike the discrete bars of histograms, density plots use a continuous curve that represents the likelihood of data points occurring at different values. The curve provides a more nuanced and visually pleasing representation of the distribution.

Area Under the Curve: The area under the density curve represents the probability of a data point falling within a specific range. The higher the curve in a certain area, the higher the likelihood of data points falling within that range.

Insight into Shape: Just like histograms, density plots show the shape of the distribution, which can reveal characteristics such as symmetry, skewness, and bimodality.

A pair plot, also known as a scatter plot matrix, is a data visualization that displays scatter plots for pairs of variables in a dataset. It's a powerful tool for exploring relationships between multiple variables simultaneously. A pair plot is particularly useful when you want to visualize how different variables interact with each other and whether there are any patterns or correlations.

Here's what pair plots show and how they work:

Scatter Plots: The main components of a pair plot are scatter plots. Each scatter plot shows the relationship between two variables, one on the x-axis and the other on the y-axis. Each combination of variables is plotted against each other, forming a grid of scatter plots.

Diagonal Plots: Along the diagonal of the pair plot grid, histograms or kernel density plots are typically displayed for each individual variable. This provides insight into the distribution of each variable.

Correlation and Patterns: Pair plots allow you to visually assess how variables are correlated with one another. If two variables have a strong linear relationship, you'll see points clustered around a line in the scatter plot.

Multivariate Exploration: Pair plots enable you to explore multivariate relationships and patterns that might not be evident when considering pairs of variables in isolation.

Identifying Outliers: Outliers, which are data points that deviate significantly from the rest of the data, can sometimes be identified in scatter plots as points that are far from the general cluster.

Symmetry: The arrangement of the scatter plots above and below the diagonal is symmetric because the same relationships are plotted twice, just with the axes swapped.

Color and Markers: If you have categorical data, you can use color or different markers to represent different categories, making it easier to identify relationships between variables and categories.

Pair plots are a powerful tool for data exploration, pattern recognition, and understanding relationships between multiple variables. They are especially useful when working with a moderate number of variables, as they provide a comprehensive overview of pairwise relationships.

In Python, you can create pair plots using libraries like seaborn and matplotlib. The seaborn.pairplot() function is commonly used for generating pair plots, and it includes options for customizing colors, markers, and more.


Box plots, also known as whisker plots, are a type of data visualization that provide insight into the distribution, spread, central tendency, and potential outliers of a dataset. They are particularly useful for comparing multiple categories or groups and identifying differences in their distributions.

Here's what box plots show and how they work:

Five-Number Summary: Box plots display the five-number summary of a dataset: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The "box" in the plot spans from Q1 to Q3, encapsulating the interquartile range (IQR).

Central Tendency: The median, represented by a horizontal line inside the box, indicates the center of the data distribution. It's also the point that divides the data into equal halves.

Spread and Dispersion: The width of the box (IQR) indicates the spread or dispersion of the middle 50% of the data. A wider box suggests higher variability.

Whiskers: Whiskers extend from the box to the minimum and maximum data points within a specified range. They can indicate the range of the data distribution and potentially identify outliers.

Outliers: Individual data points outside the whiskers are considered outliers and can be displayed individually on the plot. Outliers are data points that significantly deviate from the rest of the data and may indicate anomalies or errors.

Skewness: The relative length of the two whiskers can provide insight into the skewness of the distribution. If one whisker is longer, it suggests a skewed distribution.

Comparison: Box plots are great for comparing distributions across different categories or groups. By placing multiple box plots side by side, you can visually assess differences in central tendency, spread, and potential outliers.

Symmetry and Shape: The appearance of the box plot can provide information about the symmetry and shape of the distribution. For example, a box plot with a symmetrical box suggests a symmetric distribution.

Creating the machine learning models for the classification of wines.


```python
df = df.drop(['target_names'], axis=1)
X = df.drop(['proanthocyanins','nonflavanoid_phenols', 'malic_acid', 'alcalinity_of_ash', 'ash','alcohol', 'color_intensity', 'magnesium'], axis=1)
y = df.pop('target')
```

The first model will be SVM and we are going to test it both to scaled and unscaled data.


```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def svm_classification(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    clf = SVC(kernel='linear', C=1.0)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    confusion = confusion_matrix(y_test, y_pred)

    metrics = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    }

    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion, annot=True, cmap='Blues', fmt='g')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    return metrics


evaluation_metrics = svm_classification(X, y)
for metric, value in evaluation_metrics.items():
    print(f"{metric}: {value:.2f}")
```

We can see that we have accuracy 100% which is pretty good, although the dataset is small, due to linearity of the datapoints.


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def random_forest_classification(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    clf = RandomForestClassifier(n_estimators=100, random_state=random_state)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    confusion = confusion_matrix(y_test, y_pred)

    metrics = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1
    }

    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion, annot=True, cmap='Blues', fmt='g')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    return metrics

# Call the function to perform Random Forest classification and get evaluation metrics
evaluation_metrics = random_forest_classification(X, y)
for metric, value in evaluation_metrics.items():
    print(f"{metric}: {value:.2f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

def knn_classification(X, y, test_size=0.2, random_state=42, max_k=20):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    k_values = np.arange(1, max_k + 1)
    metrics_list = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    best_metrics = {metric: 0.0 for metric in metrics_list}
    best_model = None

    plt.figure(figsize=(8, 5))

    for k in k_values:
        clf = KNeighborsClassifier(n_neighbors=k)
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='weighted')
        recall = recall_score(y_test, y_pred, average='weighted')
        f1 = f1_score(y_test, y_pred, average='weighted')

        metrics = {
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        }

        plt.plot(k, accuracy, marker='o', label=f'k={k}')

        if accuracy > best_metrics['Accuracy']:
            best_metrics = metrics
            best_model = clf

    plt.title('KNN Classification Metrics for Different k Values')
    plt.xlabel('Number of Neighbors (k)')
    plt.ylabel('Metric Value')
    plt.legend()
    plt.show()

    return best_model, best_metrics

# Load your feature matrix (X) and target vector (y)
# For example: X = pd.DataFrame(...) and y = pd.Series(...)

# Call the function to perform KNN classification, plot metrics, and get the best model and metrics
best_knn_model, best_knn_metrics = knn_classification(X, y)

print("Best KNN Model:")
print(best_knn_model)
print("\nBest KNN Metrics:")
for metric, value in best_knn_metrics.items():
    print(f"{metric}: {value:.2f}")
```

We can see that the accuracy of KNN classifier is significantly less than 100% and that the two previous models are way better.

Last but not least we are going to check the Linear Regression model

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def linear_regression_metrics(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    metrics = {
        'Mean Squared Error': mse,
        'Root Mean Squared Error': rmse,
        'Mean Absolute Error': mae,
        'R-squared': r2
    }

    return metrics


# Call the function to perform Linear Regression and calculate evaluation metrics
evaluation_metrics = linear_regression_metrics(X, y)
for metric, value in evaluation_metrics.items():
    print(f"{metric}: {value:.2f}")
```

Again we see that the R-squared of the model is 100%.The high accuracy lays on the fact that it's a small and simple dataset yet very effective in learning the core machine learning algorithms.The approach we use can be applied to any classification problems with little modifications.
